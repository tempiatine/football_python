```python
import pandas as pd
import itertools

df = pd.read_csv('/content/sample_data/palb.csv')
df=df[['game_id', 'goal_id', 'goal_value', 'player', 'score', 'half']]

# Создаем список уникальных игроков
unique_players = df['player'].unique().tolist()
# Создаем все возможные пары игроков
pairs = list(itertools.combinations(unique_players, 2))

results = pd.DataFrame(columns=['player1', 'player2', 'goals_together', 'goals_against', 'difference'])

for pair in pairs:
    filtered_df = df[df['player'].isin(pair)]
    grouped = filtered_df.groupby('goal_id')['goal_value'].sum().reset_index()
    goals_together = grouped[grouped['goal_value'] == 2].shape[0]
    goals_against = grouped[grouped['goal_value'] == -2].shape[0]
    difference = goals_together - goals_against
    if goals_together == 0 and goals_against == 0:
      continue
    else:
      # Create a new DataFrame for the current pair
      temp_df = pd.DataFrame({'player1': [pair[0]], 'player2': [pair[1]],
                              'goals_together': [goals_together], 'goals_against': [goals_against],
                              'difference': [difference]})
      # Concatenate the new DataFrame with the results DataFrame
      results = pd.concat([results, temp_df], ignore_index=True)

results.sort_values('difference', ascending=False)

```
